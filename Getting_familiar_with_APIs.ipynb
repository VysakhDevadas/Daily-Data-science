## **API**

**WIKIPEDIA API**

```python
#lets create a query dictionary with the parameters we need from the wikipedia api
#we need action - "query" (We are performing query - which means we want the data from and about MediaWiki API)
#we need format to be "json"
#we need prop - 'revisions' - 
#we need rvprop - ' content' - rvprop is which properties to get for each revision(chosen in prop)
#Finally titles - 'Hurricane_Irene_(2005)' - we need to mention which title we want to work on (note : it can be list of titles as well)
```

```python
querydict={'action':'query', 'format':'json', 'prop':'revisions', 'rvprop':'content', 'titles':'Hurricane Irene (2005)'}
querydict
```

```python
#Now we will create a query string to and save the results in a json file
wiki ='http://en.wikipedia.org/w/api.php'
query_string = "?"+"&".join([k+"="+v for k,v in querydict.items()])
r = requests.get(wiki+query_string)
query_string
```

```python
r.json()
```

```python
r.text
```
